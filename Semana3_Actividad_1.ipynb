## **Parte #1: Fundamentos de Bases de Datos**

**Fundamentos de bases de datos para la ciencia de datos**

El flujo de trabajo de un científico de datos comienza con la definición del problema, continua con la recolección y preparación de los datos, se desarrolla el modelo, despliegue del modelo, mejora del desempeño y se evalúa si se cumple con la solución del problema.
La Base de datos, es una colección de información, se compone de una o mas listas, que se encuentran organizadas por columnas, filas y celdas llamadas tablas. Los objetos permiten obtener información de las bases de datos, que son recopilaciones compartidas de datos organizados lógicamente relacionados, con el objetivo de satisfacer las necesidades información de una organización.

Los Data Warehouse son bases de datos de gran tamaño que incorporan información de diferentes fuentes y se encuentra a disposición. Algunos de los problemas tipos que se presentan en los Data Warehouse son la normalización, datos corruptos, datos faltantes y compatibilidad entre las diferentes bases de datos que integran el Data Warehouse. Es por eso que el científico de datos necesita manejar los sistemas de administración de las bases de datos (DBMS), así como los lenguajes para manipular las bases de datos (DML). Los Data Warehouse actuales presenta ventajas como: automatización, fácil uso, seguridad, costo, y gobernanza, que los hace indispensable para el desarrollo de negocios.
