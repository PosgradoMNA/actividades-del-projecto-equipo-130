**Describe el conjunto de datos**
 Los datos contiene el registro de 1885 participantes. Para cada participante 12 atributos son conocidos: los cuales incluyen mediciones de personalidad NEO-FFI-R, BIS-11 y ImpSS, nivel de educación, edad, genero, pais de residencia y etnica. Todos los atributos son originalmente categoricos y se encuentran cuantificados. Después de la cuantificación todos los valores de las caracteristicas de entrada pueden considerarse como valores reales. 

 **¿Cuáles son algunas de las preguntas de investigación que podría responder con estos datos?**

 1._¿Consuma de drogas por genero?

 2._¿Consumo de drogas por pais y etnia?

 3._¿Qué propenso somos al consumo de drogas en base a la personalidad?

 4._¿Qué nivel educativo consume más drogas?

 5._¿Cuáles son las drogas de mayor consumo por edad?

 6._¿Es posible obtener modelos de machine learnig que predigan el consumo de drogas en personas?

 7._¿A qué edad es más frecuente iniciar en el consumo de drogas?

 8._¿Qué tipo de droga es más propensa a consumir la persona?

 9._¿Qué tipo de personalidad es más propensa al consumo de drogas?

 10._¿Qué tipo de personalidad es la menor consumo de algún tipo de droga?

 11._¿Determinar cuáles son los factores que más influyen en el consumo de drogas?

 **¿Qué actor (empresa, sector de la sociedad, organizmo descentralizado, gobierno,etc) estaría interesado en las preguntas mencionadas que identificó?**

El sector educativo desde los niveles de primaria hasta universidad estaría interesado en conocer los perfiles de los estudiantes que son propensas al consumo de drogas, basándonos en las características físicas y atributos de la personalidad, y que pudieran mermar de manera significativa el desempeño académico. Otro sector que podría estar interesado sería el sector médico, con el objetivo de tener más información del perfil del paciente, por medio del cual se obtendría mejores evaluaciones y diagnósticos clínicos. Las empresas aseguradoras de automóviles y médicas es otro sector que podría verse beneficiado al contar modelos que permitan obtener perfiles de los usuarios, para obtener primas personalizadas.
